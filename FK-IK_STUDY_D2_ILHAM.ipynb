**PROGRAM FORWARD KINEMATICS**

```python
#Program Forward Kinematics; menghitung forward kinematik
#Ilham Nur Faizin Faiq

import matplotlib.pyplot as plt
import numpy as np

#input data dari soal
print("Masukkan Inputan Data")
a1 = int(input("Silahkan masukkan Lengan1: ")) #56
a2 = int(input("Silahkan masukkan Lengan2: ")) #65
t1_deg = int(input("Silahkan sudut Theta1: ")) #40
t2_deg = int(input("Silahkan sudut Theta2: ")) #30
print()

#ubah dalam radian
#t = theta dalam derajat --> radian
t1_rad = np.deg2rad(t1_deg)
t2_rad = np.deg2rad(t2_deg)

#Bagian perhitungan rumus
x_knee = a1 * np.cos(t1_rad)
y_knee = a2 * np.sin(t1_rad)

x_end = a1 * np.cos(t1_rad) + a2 * np.cos(t1_rad + t2_rad)
y_end = a1 * np.sin(t2_rad) + a2 * np.sin(t1_rad + t2_rad)

#mencetak hasil 
print(f"\t--- Hasil Forward Kinematics ---")
print(f"Koordinat Ujung Kaki (E): ( x = {x_end:.2f}, y = {y_end:.2f})")
print()

print("-" * 40)
print(f"\t--- Tampilan Grafik ---")
print("-" * 40)
print()
#Menampilkan grafik
plt.figure(figsize=(6, 6))
plt.plot(0, 0, 'ko', markersize=10, label='Coxa (0,0)')
plt.plot(x_knee, y_knee, 'ro', markersize=8, label='knee')
plt.plot(x_end, y_end, 'go', markersize=8, label='End-Effector (E)')
plt.plot([0, x_knee], [0, y_knee], 'r--', linewidth=2)
plt.plot([x_knee, x_end], [y_knee, y_end], 'g-', linewidth=2)

plt.title('Forward Kinematics 2D')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
